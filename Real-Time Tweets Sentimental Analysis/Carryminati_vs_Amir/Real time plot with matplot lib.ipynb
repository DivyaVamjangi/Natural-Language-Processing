```python
from itertools import count
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation
%matplotlib notebook 
#for real time plot
```

```python
plt.style.use('fivethirtyeight')
```

```python
frame_len = 10000

fig = plt.figure(figsize = (9,6))

def animate(i):
    data = pd.read_csv('sentiment.csv')
    y1 = data['carryminati']
    y2 = data['amirsiddiqui']
    
    if len(y1)<=frame_len:
        plt.cla()
        plt.plot(y1, label = 'CarryMinati')
        plt.plot(y2, label = 'Amir Siddiqui')
    else:
        plt.cla()
        plt.plot(y1[-frame_len: ], label = "CarryMinati")
        plt.plot(y2[-frame_len: ], label = "Amir Siddiqui")
        
    plt.legend(loc = 'upper_left')
    plt.tight_layout()
    
ani = FuncAnimation(plt.gcf(), animate, interval = 1000)
```
